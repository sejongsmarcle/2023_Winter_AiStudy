# k겹 교차 검증
---


> 수술 전 상태와 수술 후 생존율에 대한 데이터를 k겹 교차 검증을 이용하여 예측한다.



```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold #데이터를 원하는 숫자만큼 쪼개 각각의 학습셋과 테스트셋으로 사용

import numpy as np
import tensorflow as tf
import pandas as pd

#seed 값 설정
seed=0
np.random.seed(seed)
tf.random.set_seed(seed)

#데이터 입력
df=pd.read_csv('ThoraricSurgery.csv',header=None)
```

# 1. 데이터 전처리
---


> 직접 csv 파일을 볼 필요없이 info와 head를 통해 열의 개수와 데이터의 상태를 알 수 있다. 현재 총 18 columns로 열에 맞게 데이터를 구성해주면 된다.



```python
#데이터 확인
df.info()
print()
df.head()

#데이터 전처리
dataset=df.values
X=dataset[:,0:17].astype(float)
Y=dataset[:,17]
```

# 2. k겹 교차 검증 및 모델 학습
---


```python
#3개의 파일로 쪼갬 
n_fold=3
skf= StratifiedKFold(n_splits=n_fold,shuffle=True,random_state=seed)

#교차 검증마다 accuracy 배열에 정확도 저장.
accuracy=[]

#모델 생성,컴파일,실행
for train,test in skf.split(X,Y):
  model=Sequential()
  model.add(Dense(20,input_dim=17,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(loss='binary_crossentropy',optimizer='adam',
                metrics=['accuracy'])
  model.fit(X[train],Y[train],epochs=100,batch_size=10)
  k_accuracy="%.4f"%(model.evaluate(X[test],Y[test])[1])
  accuracy.append(k_accuracy)
```

```python
#결과 출력
print("\n %.f fold_accuracy:" %n_fold,accuracy)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 6, 12, 18, 24]
y = [10, 30, 35, 50, 75]

class LinearRegression:
    def __init__(self, x, y):
        self.x_m = np.mean(x)
        self.y_m = np.mean(y)
        self.x = x
        self.y = y
        self.weight = 0
        self.bias = 0

    def forward(self):
        print(self.x, self.y)
        divisor = sum([(x - self.x_m) ** 2 for x in self.x]) # 분모
        divided = sum([(self.x[i] - self.x_m) * (self.y[i] - self.y_m) for i in range(len(self.x))])
        self.weight = divided / divisor
        self.bias = self.y_m - self.weight * self.x_m

    def predict(self):
        return [self.weight * x + self.bias for x in self.x]

model = LinearRegression(x, y)
model.forward()
y_pred = model.predict()

plt.figure(figsize=(5, 3))
plt.xlabel("SMARCLE Activity Period (month)")
plt.ylabel("major understanding (%)")
plt.scatter(x, y)
plt.plot(x, y_pred)
plt.show()
```

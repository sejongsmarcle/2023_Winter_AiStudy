# 코딩으로 확인하는 Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline

data = [[2, 0], [4, 0], [6, 0], [8, 1], [10, 1], [12, 1], [14, 1]]

x= [i[0] for i in data] # Study time
y= [i[1] for i in data] # Pass

a = 0
b = 0

lr = 0.05 # learning rate

def sigmoid(x):
    return 1 / (1 + np.e **(-x))

for i in range(2001):
    for x_data, y_data in data:
        # a에 관한 편미분
        a_diif = x_data * (sigmoid(a * x_data + b) - y_data)

        # b에 관한 편미분
        b_diff = sigmoid(a * x_data + b) - y_data

        # updata a
        a = a - lr * a_diif
        # update b
        b = b - lr * b_diff


        if i % 1000 == 0:
            print("epoch = {:.4f}, 기울기 = {:.4f}, 절편 = {:.4f}".format(i, a, b))

plt.scatter(x, y)
plt.xlim(0, 15) # range of x line
plt.ylim(-.1, 1.1) # range of y line
x_range = (np.arange(0, 15, 0.1))
plt.plot(np.arange(0, 15, 0.1), np.array([sigmoid(a * x + b) for x in x_range]))
plt.show()

```
